# Web Scraping with Beautiful Soup # 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Scrapping a Cocoa Rating Website ## 
Link to the Website <a href ="https://s3.amazonaws.com/codecademy-content/courses/beautifulsoup/cacao/index.html">Click here </a>


```python
bso=requests.get('https://s3.amazonaws.com/codecademy-content/courses/beautifulsoup/cacao/index.html')
```

```python
soup=bso.content
soup=BeautifulSoup(soup,'html.parser')
print(soup)
```

### Extracting Rating Tags ###

```python
r_s=soup.select('.Rating')
print(r_s)
print(len(r_s))
```

```python
Ratings=[]
```

```python
for rating in r_s[1:]:
    Ratings.append(float(rating.get_text()))
```

### List of Rating ###

```python
print(Ratings)
print(len(Ratings))
```

### Visualisation of Rating using Histogram ###

```python
plt.hist(Ratings)
```

```python
company=soup.select('.Company')
print(len(company))
```

```python
Companies=[]
```

```python
for comp in company[1:]:
    Companies.append(comp.get_text())
```

### List of Companies ###

```python
print(Companies)
print(len(Companies))
```

### Creating a Dictonary ### 

```python
diction={'Company':Companies,'Rating':Ratings}
df=pd.DataFrame.from_dict(diction)
```

```python
print(df.head())
```

```python
df=df.groupby('Company')
df=df[['Rating']].mean()
df=df.nlargest(10,'Rating')
```

```python
print(df.head())
```

```python
cocoa=soup.select('.CocoaPercent')
```

```python
Cocoa_Percents=[]
```

```python
for coca in cocoa[1:]:
    Cocoa_Percents.append(float(coca.get_text().strip('%')))
    
```

```python
print(Cocoa_Percents)
print(len(Cocoa_Percents))
```

```python
pdf={"Company":Companies,"Rating":Ratings,"Cocoa_Percentage":Cocoa_Percents}
pdf=pd.DataFrame(pdf)
```

```python
print(pdf.head())
```

```python
ydf=pdf.groupby('Company')
gdf=ydf[['Rating','Cocoa_Percentage']].mean()

```

```python
gdf.nlargest(10,'Rating')
```

```python
plt.scatter(gdf.Cocoa_Percentage, gdf.Rating)
z = np.polyfit(gdf.Cocoa_Percentage, gdf.Rating, 1)
line_function = np.poly1d(z)
plt.plot(gdf.Cocoa_Percentage, line_function(gdf.Cocoa_Percentage), "r--")
plt.show()
```
